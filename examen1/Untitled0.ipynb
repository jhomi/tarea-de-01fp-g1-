ejercicio 1 examen parcial 

Diseñe un algoritmo que permita determinar el nivel del perfil de ingreso de un postulante con la 
que obtiene una vacante a la carrera de ingeniería de Sistemas; considerando que, si su nota es mayor 
o igual a 17 su nivel es 4, si la nota es menor de 17 y mayor o igual a 14 su nivel es 3, si su nota es 
menor a 14 y mayor o igual a 11 su nivel es 2; mientras que si su nota es menor de 11 ya no puede 
obtener la vacante y por ende también está en el nivel 1.
Considera que la nota final es vigesimal según el siguiente ponderado:
Examen de conocimiento 40%
Entrevista Personal 35%
Test Psicológico 25%



```python
#definir variables
examenCon:float()
e:float
p:float
t:float
entrevistaPersonal:float
test:float
total:float
texto:str
#Datos de entrada
examenCon=float(input("Ingrese la nota de su examen(0-20): "))
entrevistaPersonal=float(input("Ingrese la nota de entrevista personal(0-20): "))
test=float(input("Ingrese la nota del test psicologico(0-20): "))
#Proceso
e=examenCon*0.4
t=test*0.25
p=entrevistaPersonal*0.35
total=e+p+t
if total>20:
  texto=("Introdusca de nuevo los datos")
elif total>=17:
  texto=("Su nivel es 4")
elif total>=14 and total<17:
  texto=("Su nivel es 3")
elif total>=11 and total<14:
  texto=("Su nivel es 2")
else:
  texto=("Su nivel es 1, por ende no obtiene vacante")
#Datos de salida
print(texto)
```

a pagar.

3 ejercicio 

Diseñar un algoritmo que permita calcular una operación aritmética entre 2 valores introducidos y 
el signo correspondiente por teclado: si es el signo es + debe realizar una suma, si es el signo – debe 
realizar la resta, si es el signo / debe realizar la división, si es el signo * debe realizar la multiplicación, 
si es el signo ^ debe calcular la potencia; por otro lado, si introduce R debe permitir calcular la raíz 
cuadrada de un número, si introduce % debe permitir calcular el módulo de 2.

```python
#definir varibles 
num1:float
num2:float
signo:str
total:float
#Datos de entrada
num1=float(input("Ingrese la primera  unidad: "))
num2=float(input("Ingrese la segunda unidad: "))
print("R = raiz")
print("- = resta")
print("^ = potencia")
print("% = modulo 2")
print("/ = divicion")
print("+ = suma")
print("* = multiplicacion")
signo=str(input("Ingrese el algoritmo: "))
#proceso
if signo=="+":
 total=num1+num2
elif signo=="-":
 total=num1-num2
elif signo=="/":
 total=num1/num2
elif signo=="*":
 total=num1*num2
elif signo=="^":
 total=num1**num2
elif signo=="R":
 total=num1**(num2**-1)
elif signo=="%":
 total=num1-num1
#datos de salida
print(f"El resultado es {total}")
```
